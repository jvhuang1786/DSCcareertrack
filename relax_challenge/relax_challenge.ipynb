```python
#Data Cleaning Libraries 
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.0000f' % x)
import numpy as np
import datetime
from datetime import datetime
from tqdm import tqdm
from collections import defaultdict
```

```python
takehome_users = pd.read_csv('takehome_users.csv',  encoding='iso-8859-1')
```

```python
#email feature creation
email = [x.split('@',1) for x in takehome_users['email']]
email = pd.DataFrame(email)
```

```python
split2 = [x.split('.') for x in email[1]]
split2 = pd.DataFrame(split2)
```

```python
takehome_users['email_name'] = email[0]
takehome_users['email_domain'] = split2[0]
```

```python
takehome_users['invited_by_user_id'].nunique()
takehome_users['last_session_creation_time'] = takehome_users['last_session_creation_time'].replace(np.nan, 0)
```

```python
last_time = []
for x in range(12000):
    t = int(takehome_users['last_session_creation_time'][x])
    t = datetime.datetime.fromtimestamp(t).strftime('%Y-%m-%d %H:%M:%S')
    last_time.append(t)
```

```python
takehome_users['last_session_creation_time'] = last_time
```

```python
time_diff = []
for x in range(12000):
    t = datetime.strptime(takehome_users['last_session_creation_time'][x], "%Y-%m-%d %H:%M:%S") - datetime.strptime(takehome_users['creation_time'][x], "%Y-%m-%d %H:%M:%S")
    t = t.seconds
    t = t // 60
    time_diff.append(t)
```

```python
takehome_users['time_diff_mins'] =  time_diff
```

```python
takehome_users = takehome_users.rename(columns={'object_id': 'user_id'})
```

```python
invite_user = pd.DataFrame(takehome_users['invited_by_user_id'].value_counts().reset_index())
invite_user = invite_user.rename(columns={'index':'user_id', 'invited_by_user_id':'count'})
invite_user
```

```python
data = pd.merge(takehome_users, invite_user, how = 'left', on = 'user_id')
data = data.rename(columns={'count': 'number_invited'})
data['number_invited'] = data['number_invited'].replace(np.nan,0)
```

```python
#Create a datetime date for time so we can take out the target variable login 3 out of the 7 days 
time = []
for x in tqdm(range(207917)):
    t = datetime.strptime(takehome_engage['time_stamp'][x], "%Y-%m-%d %H:%M:%S")
    t = t.strftime('%Y/%m/%d')
    time.append(t)
```

```python
takehome_engage['time'] = time
```

```python
data3 = takehome_engage
```

```python
# create a dictionary to hold all of the active users
active = defaultdict(lambda: 0)
```

```python
#for the number of entries in the users summary dataframe
for i in tqdm(np.arange(len(data3))): 
    #create a weekday variable that holds '7 days'
    wk = pd.to_timedelta('7D') 
    
    # create a variable that holds an array of all login dates for a user for each iteration
    dates = data3[data3.user_id == data3.iloc[i].user_id].time_stamp.values
    
    #if a user is in the dictionary, check if they are in the dictionary with a positive flag 
    #skip and continue with the next user to minimize duplicate computation time
    if active[data3.iloc[i].user_id] >0 : continue
    
    # for each date in the user log in array
    for date in np.arange(len(dates)):
        # set counter variable to 0
        count = 0   
        #take the date and create a range of dates 7 days long
        rng = pd.date_range(start = dates[date], end = pd.to_datetime(dates[date])+wk)
        # for each date in the user logins
        for date in dates: 
            #if the date is in the 7 day range
            if date in rng: 
                #increment the counter
                count +=1 
                 # if the counter is 3 or more
        if count >= 3:
            #set the key to the user_id and the value to 1
            active[data3.iloc[i].user_id] = 1 
```

```python
def total_visits(row, df):
    ''' function that returns the number of times visited'''
    user = row['user_id']
    return df.loc[user].visited
```

```python
# group by user id, and create a 'total_visits' feature
visits_by_user = data3.groupby('user_id').sum()
data3['ttl_visits'] = data3.apply(total_visits, df = visits_by_user, axis = 1)
```

```python
#subset the data so show only the users who visited at least 3 times
active_users = data3[data3.ttl_visits >=3]
```

```python
active_users
```

```python
def active_users_(x):
    ''' function to replace the value of a column with the active flag'''
    if int(x) in active.keys(): # the user id value is in the active dictionary keys
        return pd.Series(int(x)).replace(active)[0] #replace with the active dictionary flag
```

```python
#apply the active_users_ function to create the adopted/active users flag feature
data['active'] = data.user_id.apply(active_users_)

#fill NaN values for users that are potentially not in the active dictionary
#those users with <3 logins
data.fillna(value={'active': 0}, inplace = True)
```

```python
#Confusion matrix 
cf_matrix = confusion_matrix(y_test, y_pred)
class_names = ['Active', 'Non-Active']


def plot_cm(y_true, y_pred, figsize=(15,10)):
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=class_names, columns=class_names)
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=(15,10))
    ax = sns.heatmap(cm, cmap= "YlGnBu", annot=annot, fmt='', ax=ax)
#     bottom, top = ax.get_ylim()
#     ax.set_ylim(bottom + 0.5, top - 0.5)
    
plot_cm(y_test, y_pred)
plt.show()
```
