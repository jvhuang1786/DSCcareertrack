### Question 1- Find the 10 countries with the most project
1. Loaded the data set using pd.read_json
2. Viewed the data information by looking at df.head()
3. Looked at the column names
4. Looked at country names for 20
5. Created a dataframe for 10 countries with highest project count
    1. renamed the columns
    2. changed the index name
    3. Reset the index
6. Printed the results 

```python
#loaded the dataset using pd.read_json
df = pd.read_json('data_wrangling_json/data/world_bank_projects.json')
# View the head of df 
df.head()
#Looking at the column names
for col in df.columns: 
    print(col) 
#look at the countryname column
df.countryname.head(20)
# Use Index.value_counts() function to count the number of unique values in the given Index.
df_10 = df.countryname.value_counts().head(10)
#Creating a dataframe of the 10 countries with the highest project counts. 
country_topten = pd.DataFrame(df_10,index = None)
#index.rename(columns=) to rename the column name 
country_topten.rename(columns={'countryname':'Count'}, inplace=True)
#change the index name 
country_topten.index.name='Country Name'
#reset the index
country_topten=country_topten.reset_index()
#printing the results
print(country_topten)
```

### Question 2-Find the top 10 major project themes (using column 'mjtheme_namecode')
1. Flattened and stored the mjtheme_namecode using json.load and normalize 
2. Explored the data using .head() and .tail()
3. Filled the Nan values with NaN.  
4. Used regular expression to look for the white spaces
    1. * | Greedily matches the expression to its left 0 or more times.
    2. ^ | Matches the expression to its right at the start of a string. It matches every such instance before each \n in the string.
    3. \s | Matches whitespace characters, which include the \t, \n, \r, and space characters.
    4. $ | Matches the expression to its left at the end of a string. It matches every such instance before each \n in the string.

5. Created a duplicate data set to look at the codes to match for the project name
    1. Sometimes csv file has null values, which are later displayed as NaN in Data Frame. Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null values in different ways.
    2. Droped duplicates and sorted them
6. Turned column code to numeric using pd.to_numeric 
7. Changed project name based on code using .loc
8. Group by and sorted to see the top 10 project count

```python
# Flatten and store the mjtheme_namecode column
data = json.load((open('data_wrangling_json/data/world_bank_projects.json')))
data = json_normalize(data,'mjtheme_namecode')

#exploring the data 
data.sort_values(["code", "name"], axis=0,ascending=True, inplace=True) 

#Looking at the head and tail 
print(data.head())
print(data.tail())

#replace the blank values with NaN with regular Expression 
data=data.replace(r'^\s*$', np.nan, regex=True)

#seeing the codes that match with the project name
duplicate_data =  data.drop_duplicates(['code','name'])

#dropped duplciate values 
duplicate_data = duplicate_data.dropna(how = 'any')

#printed the results sorted the order by code then name in descending order 
print(duplicate_data.sort_values(['code','name'],ascending = True))

#Turning the code section from string to numeric 
data['code'] = pd.to_numeric(data['code'], errors='coerce')

#Changed project name based on code using .loc 
data.loc[data['code'] == 1, ['name']] = 'Economic management'
data.loc[data['code'] == 10, ['name']] = 'Rural development'
data.loc[data['code'] == 11, ['name']] = 'Environment and natural resource management'
data.loc[data['code'] == 2, ['name']] = 'Public Sector governance'
data.loc[data['code'] == 3, ['name']] = 'Rule of Law'
data.loc[data['code'] == 4, ['name']] = 'Financial and private sector development'
data.loc[data['code'] == 5, ['name']] = 'Trade and intergration'
data.loc[data['code'] == 6, ['name']] = 'Social protection and risk management'
data.loc[data['code'] == 7, ['name']] = 'Social dev/gender/inclusion'
data.loc[data['code'] == 8, ['name']] = 'Human development'
data.loc[data['code'] == 9, ['name']] = 'Urban Development'

d1 = data.groupby('name').size()
print(d1.sort_values(ascending=False).head(10))
```
